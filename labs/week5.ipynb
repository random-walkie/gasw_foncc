## Basic Histograms
Histograms also known as bar charts can be constructed from any numerical data, in this case we will work with random numbers.

First, we are going to generate a random data set of 100 floating point numbers between 0-5.

The code is as follows:


```python
import numpy
x = numpy.random.uniform(0.0, 5.0, 100)
print(x)
```

Next, we will use the random numbers to explore another feature of Python, the histogram.  Place the code below the random number code or in a cell below it.

```python
import matplotlib.pyplot as plt
plt.hist(x, 5)
plt.show()
```

## Pretty Histograms
The `hist()` method has a number of arguments that can be used to customise a histogram.  Copy the following code making sure can identify how to control the output.

```python
# x= data, bins = bars, color = RGB, alpha = transparency, rwidth = bar gap
plt.hist(x, bins=5, color='#0504aa', alpha=0.7, rwidth=0.85)
plt.grid(axis='y', alpha=0.75)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('My Very Own Histogram');
```

## Histogram using pandas data frame
It is possible to create a histogram using a panda as the data source.

```python
import pandas as pd

# Create a dictionary with sample data
data = {
    'Name': ['John', 'Emma', 'Peter', 'Olivia','Emir', 'Alia', 'Munro', 'Sally'],
    'Age': [25, 28, 32, 30, 27, 23, 31, 35],
    'City': ['New York', 'London', 'London', 'Singapore','New York', 'London', 'London', 'Paris']
}

# Create a DataFrame from the dictionary
df = pd.DataFrame(data)

dfage = df['Age']

# x= data, bins = bars, color = RGB, alpha = transparency, rwidth = bar gap
plt.hist(dfage, bins=6, color='#0504aa', alpha=0.7, rwidth=0.85)

plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution');
```

## Pie Charts
A pie chart is useful for displaying counts of category data. The following code shows a basic pie chart followed by a customised version.

### Basic Pie Chart

```python
y = [30, 15, 20, 25, 10]
z = ['Apples', 'Grapes', 'Bananas', 'Pears', 'Oranges']
plt.pie(y, labels = z)
plt.show()
```

### Pretty Pie Chart

```python
explode = [0.1,0,0,0.03,0] # To slice the sections
colors = ['#4F6272', '#B7C3F3', '#DD7596', '#8EB897', '#20AA10'] # Color of each section

plt.pie(y, labels = z, radius = 1.4, explode = explode, colors = colors, autopct = "%0.2f%%", startangle = 5)
plt.show() 
```

## Challenge – Pie Chart 
>Identify the type of data that you might find in a video streaming company and customise a pie chart to demonstrate how that data might be displayed.

```python
# Sample data
categories = [
    'TV Shows',
    'Movies',
    'Documentaries',
    'Kids Content',
    'Original Content'
]

views = np.array([450, 320, 100, 180, 250])

# Modern color palette
colors = ['#00b894', '#0984e3', '#6c5ce7', '#fdcb6e', '#e17055']

# Create figure with a light background
plt.figure(figsize=(12, 8), facecolor='white')

# Create pie chart
wedges, texts, autotexts = plt.pie(views, 
                                  labels=categories,
                                  colors=colors,
                                  autopct='%1.1f%%',
                                  pctdistance=0.85,
                                  explode=(0.02, 0.02, 0.02, 0.02, 0.02),  # Subtle separation
                                  startangle=90,
                                  textprops={'fontsize': 12})

# Add a circle for donut effect
centre_circle = plt.Circle((0,0), 0.70, fc='white')
plt.gca().add_artist(centre_circle)

# Add title
plt.title('Streaming Content Distribution', 
          pad=20,
          fontsize=16,
          fontweight='bold')

# Add total views in center
total_views = views.sum()
plt.text(0, 0, f'Total Views\n{total_views}M',
         ha='center',
         va='center',
         fontsize=14,
         fontweight='bold')

# Style percentage labels
for autotext in autotexts:
    autotext.set_fontsize(10)

# Equal aspect ratio
plt.axis('equal')

plt.show()
```

## Venn Diagrams
A Venn diagram is useful for showing the intersectional relationship between entities:

```python
from matplotlib_venn import venn2

# Use the venn2 function
venn2(subsets = (10, 5, 2), set_labels = ('Group A', 'Group B'), set_colors=('#FF0000', '#00FF00'), alpha = 0.5)
plt.show()
```

## Challenge – Venn Diagrams
>Identify the type of information that you might find in an audio streaming company and customise a Venn diagram to demonstrate how that information might be displayed.

```python
from matplotlib_venn import venn2_circles

# Create figure and axis
plt.figure(figsize=(12, 8))

# Define sets with overlapping elements
streaming_metrics = {'Play Count', 'Skip Rate', 'Completion Rate', 
                    'Time of Day', 'Session Length', 'Repeat Listens'}

user_preferences = {'Liked Songs', 'Followed Artists', 'Created Playlists',
                   'Repeat Listens', 'Favorite Genres', 'Saved Albums'}

# Create Venn diagram
venn = venn2([streaming_metrics, user_preferences], 
             set_labels=('Streaming Metrics', 'User Preferences'),
             set_colors=('#3498db', '#2ecc71'),
             alpha=0.7)

# Add title
plt.title('Audio Streaming Analytics', 
          pad=20, 
          fontsize=16, 
          fontweight='bold')

# Add circles with dashed borders
venn2_circles([streaming_metrics, user_preferences], 
              linestyle='--', 
              linewidth=1, 
              color='gray')

plt.show()
```

## Word Clouds
Word clouds are used to aid in the visualisation of word frequency:

```python
from wordcloud import WordCloud

# Create a list of word
text=("Python Python Python Java")

# Create the wordcloud object
wordcloud = WordCloud(width=480, height=480, margin=0).generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```
